```python
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
```

```python
transform = transforms.ToTensor()
```

```python
# reload the dataset for transformation
train_dataset = datasets.CIFAR100(
    root="../data",
    train=True,
    download=False,
    transform=transform
)
```

```python
image, label = train_dataset[0]
```

```python
print(type(image))
```

```python
print(type(image))
```

```python
print(image.dtype)
```

```python
print(image.min())
print(image.max())
```

```python
#we write permute(1,2,0) instead of plt.imshow(image)

# PyTorch stores images as:(C,H,W)

#Matplotlib expects:(H,W,C) .So we rearrange the dimensions: (3,32,32) permute (32,32,3)



plt.imshow(image.permute(1,2,0))
plt.title(train_dataset.classes[label])
plt.axis("off")
plt.show()
```

```python
# now we will train for more images 
from torch.utils.data import DataLoader
```

```python
train_loader = DataLoader(
    train_dataset,
    batch_size=8,
    shuffle=True
)
```

```python
#iter() think as Create a person whose job is giving me next batch
# next () : give me the next batch of data
images, labels = next(iter(train_loader))
```
