```python
import os
import shutil
import time
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# 1. Arrange Uploaded Files for PyTorch
# ---------------------------------------------------------
required_files = ['train', 'test', 'meta']
data_dir = "./data"
target_folder = os.path.join(data_dir, "cifar-100-python")
os.makedirs(target_folder, exist_ok=True)

print("--- Step 1: Organizing Uploaded Files ---")
files_ready = 0

for file in required_files:
    # If the file is just sitting in the main Colab folder, move it
    if os.path.exists(file):
        shutil.move(file, os.path.join(target_folder, file))
        print(f"✅ Moved '{file}' into the PyTorch folder.")
        files_ready += 1
    # If it's already inside the PyTorch folder from a previous run
    elif os.path.exists(os.path.join(target_folder, file)):
        print(f"✅ '{file}' is already in the correct spot.")
        files_ready += 1
    else:
        print(f"❌ Missing '{file}'! Please upload it to the Colab sidebar.")

if files_ready != 3:
    raise FileNotFoundError("Cannot proceed until 'train', 'test', and 'meta' are all uploaded.")

print("\nAll files are perfectly placed. PyTorch is ready to go!")

# ---------------------------------------------------------
# 2. Setup Device & Data
# ---------------------------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"\n--- Step 2: Setup ---")
print(f"Device: {device}")

batch_size = 256
learning_rate = 0.001
num_epochs = 15

transform_train = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.5071, 0.4865, 0.4409), (0.2673, 0.2564, 0.2762)),
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5071, 0.4865, 0.4409), (0.2673, 0.2564, 0.2762)),
])

print("\n--- Step 3: Loading Data ---")
train_dataset = torchvision.datasets.CIFAR100(root=data_dir, train=True, download=False, transform=transform_train)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)

test_dataset = torchvision.datasets.CIFAR100(root=data_dir, train=False, download=False, transform=transform_test)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

print("Success! Data is loaded and ready for model training.")
```

```python
import time

print("--- Starting Level 1: ResNet18 Training ---")
start_time = time.time()

# 1. Initialize ResNet18
resnet = torchvision.models.resnet18(weights=None, num_classes=100)
# Tweak for CIFAR-100 image size
resnet.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1, bias=False)
resnet.maxpool = nn.Identity()
resnet = resnet.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(resnet.parameters(), lr=learning_rate, weight_decay=1e-4)

# 2. Training Loop
history_resnet = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

for epoch in range(num_epochs):
    epoch_start = time.time()

    # Train
    resnet.train()
    running_loss, correct, total = 0.0, 0, 0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = resnet(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)
        correct += outputs.max(1)[1].eq(labels).sum().item()
        total += labels.size(0)

    # Validate
    resnet.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = resnet(inputs)
            val_loss += criterion(outputs, labels).item() * inputs.size(0)
            val_correct += outputs.max(1)[1].eq(labels).sum().item()
            val_total += labels.size(0)

    history_resnet['train_loss'].append(running_loss / total)
    history_resnet['train_acc'].append(100. * correct / total)
    history_resnet['val_loss'].append(val_loss / val_total)
    history_resnet['val_acc'].append(100. * val_correct / val_total)

    print(f"ResNet Epoch [{epoch+1}/{num_epochs}] | Val Acc: {history_resnet['val_acc'][-1]:.2f}% ({time.time() - epoch_start:.1f}s)")

print(f"\nResNet Training Complete in {(time.time() - start_time)/60:.2f} minutes!")

# 3. Plot ResNet Results
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history_resnet['train_loss'], label='Train Loss', marker='o')
plt.plot(history_resnet['val_loss'], label='Val Loss', marker='o')
plt.title('Level 1: ResNet18 Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history_resnet['train_acc'], label='Train Acc', marker='o')
plt.plot(history_resnet['val_acc'], label='Val Acc', marker='o')
plt.title('Level 1: ResNet18 Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import time
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

print("--- Starting Level 1: ResNet18 Training (Optimized) ---")
start_time = time.time()

# Ensure device is set (assuming device is already defined in your notebook, but defining here to be safe)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- USER REQUESTED CHANGES ---
batch_size = 128
num_epochs = 50 # Increased from 15
learning_rate = 0.001
early_stop_patience = 7 # Wait 7 epochs before stopping if no improvement
# ------------------------------

# --- 1. Data Pipeline (Batch Size = 128) ---
data_dir = "./data"

transform_train = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.5071, 0.4865, 0.4409), (0.2673, 0.2564, 0.2762)),
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5071, 0.4865, 0.4409), (0.2673, 0.2564, 0.2762)),
])

train_dataset = torchvision.datasets.CIFAR100(root=data_dir, train=True, download=True, transform=transform_train)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)

test_dataset = torchvision.datasets.CIFAR100(root=data_dir, train=False, download=True, transform=transform_test)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)


# --- 2. Initialize ResNet18 ---
resnet = torchvision.models.resnet18(weights=None, num_classes=100)
# Tweak for CIFAR-100 image size (32x32)
resnet.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1, bias=False)
resnet.maxpool = nn.Identity()
resnet = resnet.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(resnet.parameters(), lr=learning_rate, weight_decay=1e-4)

# Added a scheduler to drop learning rate when it plateaus, crucial for training from scratch over many epochs
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)

# --- 3. Training Loop with Early Stopping ---
history_resnet = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': [], 'lr': []}
best_val_loss = float('inf')
counter = 0
best_epoch = 0

for epoch in range(num_epochs):
    epoch_start = time.time()

    # Train
    resnet.train()
    running_loss, correct, total = 0.0, 0, 0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = resnet(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        correct += outputs.max(1)[1].eq(labels).sum().item()
        total += labels.size(0)

    # Validate
    resnet.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = resnet(inputs)
            val_loss += criterion(outputs, labels).item() * inputs.size(0)
            val_correct += outputs.max(1)[1].eq(labels).sum().item()
            val_total += labels.size(0)

    # Metrics
    epoch_train_loss = running_loss / total
    epoch_train_acc = 100. * correct / total
    epoch_val_loss = val_loss / val_total
    epoch_val_acc = 100. * val_correct / val_total
    current_lr = optimizer.param_groups[0]['lr']

    history_resnet['train_loss'].append(epoch_train_loss)
    history_resnet['train_acc'].append(epoch_train_acc)
    history_resnet['val_loss'].append(epoch_val_loss)
    history_resnet['val_acc'].append(epoch_val_acc)
    history_resnet['lr'].append(current_lr)

    print(f"ResNet Epoch [{epoch+1}/{num_epochs}] | LR: {current_lr:.6f} | Val Loss: {epoch_val_loss:.4f} | Val Acc: {epoch_val_acc:.2f}% ({time.time() - epoch_start:.1f}s)")

    # Step the scheduler based on validation loss
    scheduler.step(epoch_val_loss)

    # --- EARLY STOPPING LOGIC ---
    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        best_epoch = epoch + 1
        counter = 0
        torch.save(resnet.state_dict(), 'best_resnet18_scratch.pth')
    else:
        counter += 1
        if counter >= early_stop_patience:
            print(f"\n🛑 EARLY STOPPING TRIGGERED! Validation loss stopped improving. Peak reached at Epoch {best_epoch}.")
            break

print(f"\nResNet Training Complete in {(time.time() - start_time)/60:.2f} minutes!")
print(f"🏆 Ultimate Peak Validation Accuracy: {max(history_resnet['val_acc']):.2f}%")

# --- 4. Plot ResNet Results ---
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.plot(history_resnet['train_loss'], label='Train Loss', marker='o')
plt.plot(history_resnet['val_loss'], label='Val Loss', marker='o')
plt.axvline(x=best_epoch-1, color='black', linestyle='--', label='Optimal Peak')
plt.title('Level 1: ResNet18 Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(history_resnet['train_acc'], label='Train Acc', marker='o')
plt.plot(history_resnet['val_acc'], label='Val Acc', marker='o')
plt.axvline(x=best_epoch-1, color='black', linestyle='--', label='Optimal Peak')
plt.title('Level 1: ResNet18 Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(history_resnet['lr'], label='Learning Rate', color='purple', drawstyle='steps-post')
plt.title('Learning Rate Tracking')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import time
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

print("--- Starting ULTIMATE Level 2: EfficientNet-B0 (Transfer Learning) ---")
start_time = time.time()

# =====================================================================
# 1. THE ADVANCED DATA PIPELINE (Upscaling & Heavy Augmentation)
# =====================================================================
print("Preparing 224x224 image pipeline for pre-trained EfficientNet...")

transform_train_eff = transforms.Compose([
    transforms.Resize(224), # Upscale to match ImageNet resolution
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(224, padding=4, padding_mode='reflect'),
    transforms.ToTensor(),
    # Using the standard ImageNet normalization values since we are using pre-trained weights
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

transform_test_eff = transforms.Compose([
    transforms.Resize(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Reloading datasets with the new 224x224 transformation pipeline
train_dataset_eff = torchvision.datasets.CIFAR100(root=data_dir, train=True, download=False, transform=transform_train_eff)
train_loader_eff = torch.utils.data.DataLoader(train_dataset_eff, batch_size=64, shuffle=True, num_workers=2)

test_dataset_eff = torchvision.datasets.CIFAR100(root=data_dir, train=False, download=False, transform=transform_test_eff)
test_loader_eff = torch.utils.data.DataLoader(test_dataset_eff, batch_size=64, shuffle=False, num_workers=2)

# =====================================================================
# 2. INITIALIZE TRANSFER LEARNING MODEL
# =====================================================================
# Load pre-trained weights from ImageNet
weights = torchvision.models.EfficientNet_B0_Weights.DEFAULT
effnet = torchvision.models.efficientnet_b0(weights=weights)

# Modify the final classification layer to output 100 classes instead of 1000
in_features = effnet.classifier[1].in_features
effnet.classifier[1] = nn.Linear(in_features, 100)
effnet = effnet.to(device)

# =====================================================================
# 3. HYPERPARAMETERS, OPTIMIZER, & SCHEDULER
# =====================================================================
max_epochs = 25 # Transfer learning converges much quicker; 25 epochs is plenty
early_stop_patience = 4
best_val_loss = float('inf')
counter = 0
best_epoch = 0

criterion = nn.CrossEntropyLoss()

# Start with a safe, stable learning rate for fine-tuning pre-trained features
optimizer = optim.Adam(effnet.parameters(), lr=0.0001, weight_decay=1e-4)

# Dynamic Scheduler to drop learning rate when accuracy plateaus
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

# =====================================================================
# 4. TRAINING & VALIDATION LOOP WITH EARLY STOPPING
# =====================================================================
history_effnet = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': [], 'lr': []}

for epoch in range(max_epochs):
    epoch_start = time.time()

    # Train Phase
    effnet.train()
    running_loss, correct, total = 0.0, 0, 0
    for inputs, labels in train_loader_eff:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = effnet(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)
        correct += outputs.max(1)[1].eq(labels).sum().item()
        total += labels.size(0)

    # Validation Phase
    effnet.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for inputs, labels in test_loader_eff:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = effnet(inputs)
            batch_loss = criterion(outputs, labels).item() * inputs.size(0)
            val_loss += batch_loss
            val_correct += outputs.max(1)[1].eq(labels).sum().item()
            val_total += labels.size(0)

    # Calculate Epoch Metrics
    current_train_loss = running_loss / total
    current_train_acc = 100. * correct / total
    current_val_loss = val_loss / val_total
    current_val_acc = 100. * val_correct / val_total
    current_lr = optimizer.param_groups[0]['lr']

    history_effnet['train_loss'].append(current_train_loss)
    history_effnet['train_acc'].append(current_train_acc)
    history_effnet['val_loss'].append(current_val_loss)
    history_effnet['val_acc'].append(current_val_acc)
    history_effnet['lr'].append(current_lr)

    print(f"EffNet Epoch [{epoch+1}/{max_epochs}] | LR: {current_lr:.6f} | Val Loss: {current_val_loss:.4f} | Val Acc: {current_val_acc:.2f}% ({time.time() - epoch_start:.1f}s)")

    # Step the scheduler based on validation loss performance
    scheduler.step(current_val_loss)

    # --- EARLY STOPPING CHECK ---
    if current_val_loss < best_val_loss:
        best_val_loss = current_val_loss
        best_epoch = epoch + 1
        counter = 0
        # Save the absolute best weights configuration
        torch.save(effnet.state_dict(), 'best_efficientnet_transfer.pth')
    else:
        counter += 1
        if counter >= early_stop_patience:
            print(f"\n🛑 EARLY STOPPING TRIGGERED! Performance stabilized at Epoch {best_epoch}.")
            break

print(f"\n✅ EfficientNet Training Complete in {(time.time() - start_time)/60:.2f} minutes!")
print(f"🏆 Ultimate Peak Validation Accuracy: {max(history_effnet['val_acc']):.2f}%")

# =====================================================================
# 5. METRICS VISUALIZATION
# =====================================================================
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.plot(history_effnet['train_loss'], label='Train Loss', marker='o', color='orange')
plt.plot(history_effnet['val_loss'], label='Val Loss', marker='o', color='red')
plt.axvline(x=best_epoch-1, color='black', linestyle='--', label='Optimal Peak')
plt.title('Ultimate Level 2: Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(history_effnet['train_acc'], label='Train Acc', marker='o', color='orange')
plt.plot(history_effnet['val_acc'], label='Val Acc', marker='o', color='red')
plt.axvline(x=best_epoch-1, color='black', linestyle='--', label='Optimal Peak')
plt.title('Ultimate Level 2: Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(history_effnet['lr'], label='Learning Rate', color='purple', drawstyle='steps-post')
plt.title('Learning Rate Tracking')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import time
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

print("--- Starting Level 2: ResNet50 Training (Optimized) ---")
start_time = time.time()

# Ensure device is set
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

if device.type == 'cpu':
    print("\n⚠️ WARNING: NO GPU DETECTED! Training ResNet50 on CPU will be extremely slow.")

batch_size = 128
num_epochs = 100 # ResNet50 is deep; it needs more epochs to converge from scratch
learning_rate = 0.1

# --- 1. Data Pipeline ---
data_dir = "./data"

transform_train = transforms.Compose([
    transforms.TrivialAugmentWide(),
    transforms.RandomCrop(32, padding=4, padding_mode='reflect'),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.5071, 0.4865, 0.4409), (0.2673, 0.2564, 0.2762)),
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5071, 0.4865, 0.4409), (0.2673, 0.2564, 0.2762)),
])

# Downloading the data securely
train_dataset = torchvision.datasets.CIFAR100(root=data_dir, train=True, download=True, transform=transform_train)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)

test_dataset = torchvision.datasets.CIFAR100(root=data_dir, train=False, download=True, transform=transform_test)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)

# --- 2. Initialize ResNet50 ---
# weights=None explicitly tells PyTorch to initialize random weights (From Scratch)
resnet = torchvision.models.resnet50(weights=None, num_classes=100)

# CRITICAL TWEAK FOR CIFAR-100 (32x32 images)
# ResNet50's default first layer is designed for 224x224 ImageNet images (7x7 conv, stride 2).
# If we don't change this, it will compress our tiny 32x32 images into useless 1x1 feature maps.
resnet.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1, bias=False)
resnet.maxpool = nn.Identity()
resnet = resnet.to(device)

criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
optimizer = optim.SGD(resnet.parameters(), lr=learning_rate, momentum=0.9, weight_decay=5e-4)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)

# --- 3. Training Loop ---
history_resnet = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': [], 'lr': []}
best_val_loss = float('inf')
best_epoch = 0

for epoch in range(num_epochs):
    epoch_start = time.time()

    # Train
    resnet.train()
    running_loss, correct, total = 0.0, 0, 0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device, non_blocking=True), labels.to(device, non_blocking=True)

        optimizer.zero_grad()
        outputs = resnet(inputs)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        correct += outputs.max(1)[1].eq(labels).sum().item()
        total += labels.size(0)

    # Validate
    resnet.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device, non_blocking=True), labels.to(device, non_blocking=True)

            outputs = resnet(inputs)
            val_loss += criterion(outputs, labels).item() * inputs.size(0)
            val_correct += outputs.max(1)[1].eq(labels).sum().item()
            val_total += labels.size(0)

    # Metrics
    epoch_train_loss = running_loss / total
    epoch_train_acc = 100. * correct / total
    epoch_val_loss = val_loss / val_total
    epoch_val_acc = 100. * val_correct / val_total
    current_lr = optimizer.param_groups[0]['lr']

    history_resnet['train_loss'].append(epoch_train_loss)
    history_resnet['train_acc'].append(epoch_train_acc)
    history_resnet['val_loss'].append(epoch_val_loss)
    history_resnet['val_acc'].append(epoch_val_acc)
    history_resnet['lr'].append(current_lr)

    print(f"ResNet50 Epoch [{epoch+1}/{num_epochs}] | LR: {current_lr:.6f} | Val Loss: {epoch_val_loss:.4f} | Val Acc: {epoch_val_acc:.2f}% ({time.time() - epoch_start:.1f}s)")

    scheduler.step()

    # Save best model based on validation loss
    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        best_epoch = epoch + 1
        torch.save(resnet.state_dict(), 'best_resnet50_scratch.pth')

print(f"\nResNet50 Training Complete in {(time.time() - start_time)/60:.2f} minutes!")
print(f"🏆 Ultimate Peak Validation Accuracy: {max(history_resnet['val_acc']):.2f}%")

# --- 4. Plot ResNet Results ---
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.plot(history_resnet['train_loss'], label='Train Loss', marker='o', color='royalblue')
plt.plot(history_resnet['val_loss'], label='Val Loss', marker='o', color='darkorange')
plt.axvline(x=best_epoch-1, color='black', linestyle='--', label='Optimal Peak')
plt.title('Level 2: ResNet50 Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(history_resnet['train_acc'], label='Train Acc', marker='o', color='royalblue')
plt.plot(history_resnet['val_acc'], label='Val Acc', marker='o', color='darkorange')
plt.axvline(x=best_epoch-1, color='black', linestyle='--', label='Optimal Peak')
plt.title('Level 2: ResNet50 Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(history_resnet['lr'], label='Learning Rate', color='purple')
plt.title('Cosine Learning Rate Curve')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
