```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from  sklearn import metrics
```

```python
house_price_dataset= sklearn.datasets.fetch_california_housing()
```

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data,columns = house_price_dataset.feature_names )
```

```python
house_price_dataframe.head()
```

```python
#add target columns
house_price_dataframe['budget'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#checkin the number of rows and columns
house_price_dataframe.shape
```

```python
#check for missing value
house_price_dataframe.isnull().sum()
```

```python
#statiscical measures
house_price_dataframe.describe()
```

```python
#understanding the  corelation
correlation = house_price_dataframe.corr()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar =True ,square= True,fmt ='.1f',annot=True, annot_kws={'size':8},cmap='RdPu')
```

```python
#splitting
X= house_price_dataframe.drop(['budget'],axis=1)
Y=house_price_dataframe['budget']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test =train_test_split(X,Y, test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#model traing(xgboost ) here we will use ensemble learning

model =XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
#accuracy on taring data
training_data_prediction = model.predict(X_train)
```

```python
#R square
score_1 = metrics.r2_score(Y_train,training_data_prediction )
score_2 =metrics.mean_absolute_error(Y_train,training_data_prediction)
print("R squared errror: ",score_1)
print("mean absolute eeroor: ",score_2)
```

```python
#visulaize
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("actualbuget")
plt.ylabel("predicted buget")
plt.title("actual vs predicted")
plt.show()
```

```python
#prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
#R square
score_1 = metrics.r2_score(Y_test,test_data_prediction )
score_2 =metrics.mean_absolute_error(Y_test,test_data_prediction)
print("R squared errror: ",score_1)
print("mean absolute eeroor: ",score_2)
```
