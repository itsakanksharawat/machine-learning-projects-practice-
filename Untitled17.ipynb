```python
import numpy as np

# example dataset: 3 features
X = np.array([
    [1, 2, 3],
    [2, 1, 0],
    [3, 4, 1],
    [4, 3, 2]
])


y = 2*X[:,0] + 1*X[:,1] + 3*X[:,2] + 5


n_features = X.shape[1]
w = np.zeros(n_features)
b = 0.0

lr = 0.01
epochs = 10

for epoch in range(epochs):
    for x, target in zip(X, y):
        y_pred = np.dot(w, x) + b
        error = target - y_pred


        w += lr * x * error
        b += lr * error

print("Learned weights:", w)
print("Learned bias:", b)
```
